
##  Verification des valeurs manquantes

On peut  avoir deux types de valeurs manquantes possibles avec ce dataset  :
1. Valeurs **NULL** (manquantes reelles)
2. Valeurs **"unknown"** (manquantes déguisées dans les variables catégorielles)

```python
# Verif des valeurs NULL

missing_values = df.isnull().sum()
missing_percent = (missing_values / len(df)) * 100

missing_df = pd.DataFrame({
    'Nombre_manquantes': missing_values,
    'Pourcentage': missing_percent
})

```

```python
missing_df = missing_df[missing_df['Nombre_manquantes'] > 0].sort_values(
    'Nombre_manquantes', ascending=False
)

if len(missing_df) > 0:
    print(" Valeurs NULL détectées :")
    print(missing_df)
else:
    print(" Pas de valeurs  NULL détectées dans le dataset ")
   
```

```python
# Verif des valeurs "unknown" dans les variables catégorielles

unknown_counts = {}
for col in categorical_cols:
    unknown_count = (df[col] == 'unknown').sum()
    if unknown_count > 0:
        unknown_counts[col] = {
            'count': unknown_count,
            'percentage': round((unknown_count / len(df)) * 100, 2)
        }

if unknown_counts:
    unknown_df = pd.DataFrame(unknown_counts).T
    unknown_df = unknown_df.sort_values('count', ascending=False)
    print("\n Variables contenant des 'unknown' sont  :\n")
    print(unknown_df)
    

    print("  Ces 'unknown' représentent des valeurs manquantes deguisées que l'on doit soit garder, supprimer ...")
    
else:
    print(" Aucune valeur 'unknown' détectee dans le dataset ")
```



##  Analyse de la Variable Cible 



```python
target_counts = df['y'].value_counts()
target_percent = (target_counts / len(df)) * 100

result_df = pd.DataFrame({
    'Effectif': target_counts,
    'Pourcentage': target_percent.round(2)
})

print("\n")
print(result_df)


print(" Observations")

print(f"\n    Taux de souscription (yes) : {target_percent['yes']:.2f}%")
print(f"    Taux de refus (no) : {target_percent['no']:.2f}%")
print(f"     Ratio de déséquilibre : {target_counts['no'] / target_counts['yes']:.1f}:1")

```

Créons deux  graphiques pour mieux comprendre visualiser 

```python
# Visualisation de la variable cible
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Graphique 1 : Barplot
sns.countplot(data=df, x='y', ax=axes[0], palette='Set2')
axes[0].set_title('Distribution de la Variable Cible (y)', 
                   fontsize=14 , fontweight='bold')
axes[0].set_xlabel('Souscription au dépôt à terme', fontsize=12)
axes[0].set_ylabel('Nombre de clients', fontsize=12)

# Ajout  des valeurs sur les barres
for container in axes[0].containers:
    axes[0].bar_label(container, fmt='%d')

    

# Graphique 2 : Pie chart
colors = ["#F2EB24", "#da0e0e"]
axes[1].pie(target_counts, labels=target_counts.index, autopct='%1.1f%%', 
            startangle=90, colors=colors, textprops={'fontsize': 12})
axes[1].set_title('Proportion de Souscription', 
                   fontsize=14, fontweight='bold')


plt.show()

plt.tight_layout()
plt.savefig(r'A:\Mes documents\projet analyse exploiratoire\results\figures\01_distribution_variable_cible.png', 
            dpi=300, bbox_inches='tight')

print(" Graphique sauvegardé dans  : results/figures/01_distribution_variable_cible.png")
```

```python
# Distribution de l'age des clients
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Graphique 1 : Histogramme
axes[0].hist(df['age'], bins=50, edgecolor='black', color='skyblue', alpha=0.7)
axes[0].set_title("Distribution de l'Âge des Clients", 
                   fontsize=14, fontweight='bold')
axes[0].set_xlabel('Âge', fontsize=12)
axes[0].set_ylabel('Fréquence', fontsize=12)
axes[0].axvline(df['age'].mean(), color='red', linestyle='--', 
                linewidth=2, label=f'Moyenne: {df["age"].mean():.1f}')
axes[0].axvline(df['age'].median(), color='green', linestyle='--', 
                linewidth=2, label=f'Médiane: {df["age"].median():.1f}')
axes[0].legend()

# Graphique 2 : Boxplot
sns.boxplot(data=df, y='age', ax=axes[1], color='lightcoral')
axes[1].set_title("Boxplot de l'Âge", fontsize=14, fontweight='bold')
axes[1].set_ylabel('Âge', fontsize=12)
axes[1].grid(axis='x', alpha=0.3)

plt.tight_layout()


plt.savefig(r'A:\Mes documents\projet analyse exploiratoire\results\figures\02_distribution_age.png', 
            dpi=300, bbox_inches='tight')
plt.show()

print(" Graphique sauvegardé : results/figures/02_distribution_age.png")

# Statistiques de l'âge
print(f"\n Statistiques de l'âge :")
print(f"   - Âge moyen : {df['age'].mean():.1f} ans")
print(f"   - Âge médian : {df['age'].median():.1f} ans")
print(f"   - Âge minimum : {df['age'].min()} ans")
print(f"   - Âge maximum : {df['age'].max()} ans")
print(f"   - Écart-type : {df['age'].std():.1f} ans")
```
